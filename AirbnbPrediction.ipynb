```python
# Import Libraries
import numpy as np
import pandas as pd
import datetime as dt
import seaborn as sns
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import MultiLabelBinarizer 
from sklearn.ensemble import RandomForestRegressor
from sklearn import metrics
from sklearn.model_selection import train_test_split
import warnings
from pandas.core.common import SettingWithCopyWarning
from sklearn.ensemble import RandomForestRegressor
from sklearn import metrics
import matplotlib.pyplot as plt

warnings.simplefilter(action="ignore", category=SettingWithCopyWarning)
mlb = MultiLabelBinarizer()
scaler = StandardScaler()
```

### Helper functions

Multiple functions are assigned to follow the DRY principle and show details of the dataframes.

```python
### Initiate Helperfunctions
def percent_to_numeric(df, col, fill=False):
    """
    Description: This function gets rid of $ or % signs and transforms the column from string to numeric

    Arguments:
        df: DataFrame Object 
        col: The column name to transform
        fill: If true, mean inputation is performed

    Returns:
        transformed column
    """
    df[col] = df[col].replace("\\$", "", regex=True)
    df[col] = df[col].replace("\\%", "", regex=True)
    df[col] = df[col].replace("\\,", "", regex=True)
    df[col] = pd.to_numeric(df[col], errors="raise")
    if fill:
        df[col] = df[col].fillna(df[col].mean())
    return df[col]
    
def missing_values_table(df):
        """
        Description: Creates a table with 

        Arguments:
            df: DataFrame Object 

        Returns:
            DataFrame with name of missing values and % of missing values
        """
        mis_val = df.isnull().sum()
        mis_val_percent = 100 * df.isnull().sum() / len(df)
        mis_val_table = pd.concat([mis_val, mis_val_percent], axis=1)
        mis_val_table_ren_columns = mis_val_table.rename(
        columns = {0 : 'Missing Values', 1 : '% of Total Values'})
        mis_val_table_ren_columns = mis_val_table_ren_columns[
            mis_val_table_ren_columns.iloc[:,1] != 0].sort_values(
        '% of Total Values', ascending=False).round(1)
        print ("Your selected dataframe has " + str(df.shape[1]) + " columns.\n"      
            "There are " + str(mis_val_table_ren_columns.shape[0]) +
              " columns that have missing values.")
        return mis_val_table_ren_columns
    
def create_named_dummy(col):
    """
    Description: Creates dummy variables for a categorical column

    Arguments:
        col: Categorical variable / column

    Returns:
        DataFrame with dummy variables
    """
    return pd.get_dummies(df[col], drop_first=True).rename(columns=lambda x: col + "_" + str(x))

def room_categories(number):
    """
    Description: Converts the room categories variable from numierc to categorical

    Arguments:
        number: Number to convert into one of three categories

    Returns:
        DataFrame with dummy variables
    """
    if(number > 0 and number < 4):
        return "small"
    elif(number > 4 and number < 6):
        return "medium"
    else: 
        return "hard"

def clean_ameniti(string):
    """
    Description: removes brackets or comata from string 

    Arguments:
        string: String to remove elements from

    Returns:
        cleaned string
    """
    return string.replace('{', '').replace('}', '').replace(" ", "").replace('"', '')
```

## Data Understanding
We start with importing the datasets. The model will be built with the listings dataset, but also consider the calendar datasets, which shall provide a descriptive analysis for seasonal fluctuation.

```python
# Read and combine datasets
calendar_boston = pd.read_csv("./data/boston/calendar.csv")
calendar_boston["city"] = "Boston"
listings_boston = pd.read_csv("./data/boston/listings.csv")
listings_boston["city"] = "Boston"
reviews_boston = pd.read_csv("./data/boston/reviews.csv")
reviews_boston["city"] = "Boston"

calendar_seattle = pd.read_csv("./data/seattle/calendar.csv")
calendar_seattle["city"] = "Seattle"
listings_seattle = pd.read_csv("./data/seattle/listings.csv")
listings_seattle["city"] = "Seattle"
reviews_seattle = pd.read_csv("./data/seattle/reviews.csv")
reviews_seattle["city"] = "Seattle"

# Concatenate all datasets 
calendar = pd.concat([calendar_boston, calendar_seattle])
listings = pd.concat([listings_boston, listings_seattle])
reviews = pd.concat([reviews_boston, reviews_seattle])
```

The column "price" is a string and has to be converted to a numeric variable

```python
# Group dataset by month and city and check how it changes over the year
calendar["price"] = percent_to_numeric(calendar, "price")
calendar["date"] = pd.to_datetime(calendar["date"])
calendar["month"] = calendar["date"].dt.month
  
mean_price_per_month = calendar.groupby(["city", "month"]).mean()["price"].reset_index()

sns.lineplot(x="month", y="price", hue="city", data=mean_price_per_month)
```

#### Listings Dataset
Now we analyze the "listings" dataset. Only a fraction of the data really makes sense for a prediction model, therefore the data set is significantly reduced.

```python
# Only keep relevant variables in dataset
df = listings[["price", "host_response_rate", "host_acceptance_rate", "host_is_superhost",
               "host_listings_count", "property_type","room_type", "accommodates", "bathrooms", "bedrooms", 
               "beds", "bed_type", "amenities", "review_scores_rating", "cancellation_policy", 
               "reviews_per_month", "city"]]
df.shape
```

```python
# Check for missing values
missing_values_table(df)
```

```python
# Preprocess dataset: Fill NA´s, convert price to numeric
df["price"] = percent_to_numeric(df, "price", fill=False)
df["host_response_rate"] = percent_to_numeric(df, "host_response_rate", fill=False)
df["host_acceptance_rate"] = percent_to_numeric(df, "host_acceptance_rate", fill=False)
```

#### Dealing with missing values
We can see that a lot of columns have missing values. 
4 variables have a lot of missing values, so we should take a look at the variability to check whether the rows should be dropped or imputed. If the columns are skewed, we would prefer a mode imputation over a mean or median imputation.

```python
fig, ax = plt.subplots(2,2)
sns.histplot(df["review_scores_rating"], ax=ax[0, 0])
sns.histplot(df["reviews_per_month"], ax=ax[0, 1])
sns.histplot(df["host_acceptance_rate"], ax=ax[1, 0])
sns.histplot(df["host_response_rate"], ax=ax[1, 1])
fig.show()
```

#### Explanation for where variables are dropped and why some are replaced
The rows of bathrooms, bedrooms, beds, property_type, host_is_superhost and host_listings_count are very few, so dropping them will not impact the model performance. 
host_response_rate, host_acceptance_rate, reviews_per_month and review_scores_rating have 13,4% to 19,7% missing values. Just dropping them could definitely harm the model performance. When we take a look at the distribution of the values, we can see that all 4 variables are skewed. We use mode imputation in that case, because the mode presents the highest probability for the missing value to be exactly that value. 

```python
df.dropna(subset=["property_type", "bathrooms", "bedrooms", "beds", "host_is_superhost", "host_listings_count"], axis=0, inplace=True)
df["review_scores_rating"] = df["review_scores_rating"].fillna(df["review_scores_rating"].mode()[0])
df["reviews_per_month"] = df["reviews_per_month"].fillna(df["reviews_per_month"].mode()[0])
df["host_acceptance_rate"] = df["host_acceptance_rate"].fillna(df["host_acceptance_rate"].mode()[0])
df["host_response_rate"] = df["host_response_rate"].fillna(df["host_response_rate"].mode()[0])
df.isna().any().any()
```

#### After getting rid of missing values, it is time to create the final dataset. We have to do the following steps:

* Numeric values: Scaling: This is done to normalize the data within a particular range, which gives all variables equal importance in the initial model
* Categorical variables: We have to build dummy varibales, since categories have no distinct "distance" between their cateogies (dropping first to prevent multicollinearity).

The "accommodates" variable gets recoded into a dummy variable to reduce variability in the column, since there are a few outliers, which could lead to an overfitted model.

```python
# Scale numeric variables, dummy variables for categorical variables
numeric_vars = pd.DataFrame(scaler.fit_transform(df.select_dtypes("float64").to_numpy()), columns = df.select_dtypes("float64").columns)

superhost_dummy = pd.get_dummies(df["host_is_superhost"], drop_first=True, prefix="superhost")
df["property_type"] = np.where(df["property_type"].map(df["property_type"].value_counts()) > 50, df["property_type"], "Other")

property_dummies = pd.get_dummies(df["property_type"], drop_first=True, prefix="property")
df["room_type"] = np.where(df["room_type"].map(df["room_type"].value_counts()) > 50, df["room_type"], "Other")
room_dummies = pd.get_dummies(df["room_type"], drop_first=True, prefix="room")

df["accommodates"] = df["accommodates"].apply(lambda x: room_categories(x))
accomodates_dummies = pd.get_dummies(df["accommodates"], drop_first=True, prefix="accomodates")

bed_dummies = pd.get_dummies(df["bed_type"], drop_first=True, prefix="bed")

df["amenities"] = df["amenities"].apply(lambda x: clean_ameniti(x))
amenities = df["amenities"].apply(lambda x: x.split(','))
amenities = pd.DataFrame(mlb.fit_transform(amenities),columns=mlb.classes_, index=df.index)
amenities.drop([""], inplace=True, axis=1)

cancellation_dummies = pd.get_dummies(df["cancellation_policy"], drop_first=True, prefix="accomodates")
city_dummies = pd.get_dummies(df["city"], drop_first=True, prefix="city")

dfs = [df.reset_index() for df in [numeric_vars, superhost_dummy, property_dummies, room_dummies, accomodates_dummies, bed_dummies, amenities, cancellation_dummies, city_dummies]]
final_df = pd.concat(dfs, axis = 1).drop(columns=["index"])
```

## Model data

1. Use full dataset
2. Drop single column (backward elimination) to reduce model
3. Plot the results to look for a threshold which variables to include and which to drop

```python
# Use Backward elimination
model_df = final_df.copy()

mapes = []
mape_df = pd.DataFrame(columns = ["Column", "MAPE"])

for index, column in enumerate(list(range(model_df.shape[1]))):
    if model_df.shape[1] == 1: # Prevent ValueError: at least one array or dtype is required
        break
    X = model_df.iloc[:, 1:model_df.shape[1]]
    y = model_df.iloc[:, 0].values
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
    regressor = RandomForestRegressor(n_estimators=100, random_state=0)
    regressor.fit(X, y)
    pred = regressor.predict(X_test)
    feat_importances = pd.Series(regressor.feature_importances_, index=X_train.columns)
    mapes.append(metrics.mean_absolute_error(y_test, pred))
    print('Mean Absolute Error (MAE):', metrics.mean_absolute_error(y_test, pred))
    print('Columns in dataset: ', model_df.shape[1])
    print('Column dropped: ', feat_importances.sort_values().reset_index().iloc[0,0]) # Drop column with worst features
    newrow = pd.DataFrame([[feat_importances.sort_values().reset_index().iloc[0,0], metrics.mean_absolute_error(y_test, pred)]], columns=["Column", "MAPE"])
    mape_df = pd.concat([mape_df, newrow], ignore_index=True)
    model_df.drop(columns=[feat_importances.sort_values().reset_index().iloc[0,0]], inplace=True)
```

```python
# Check plot to see how Error changes when least important variable was removed
mape_df["MAPE"].plot()
mape_df.sort_values(by="MAPE", ascending=False).head(10)
```

```python
# Create a final dataset to be able to work with
### Get final columns
final_model_columns = mape_df[mape_df["MAPE"] > 0.15]["Column"]
final_model_df = pd.concat([final_df["price"], final_df[final_model_columns]], axis=1)
final_model_df
```

```python
X = final_model_df.iloc[:, 1:final_model_df.shape[1]]
y = final_model_df.iloc[:, 0].values
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
regressor = RandomForestRegressor(n_estimators=100, random_state=0)
regressor.fit(X, y)
pred = regressor.predict(X_test)
resultcheck = pd.concat([pd.Series(pred), pd.Series(y_test)])
sns.barplot(data=resultcheck)
```

When we take the difference of the prediction and the actual value, we can see that the reduced model still perfoems quite well.

## Answer 1

## Answer 2: 
Some variables might influece the price as expected, like bedrooms or bathrooms. Other variables like "reviews per month", might be expected to have less influece. But these "soft" factors show social proof for the quality and therefore people seem to be willing to pay more. 
## Answer 3: 
One factor is very important: The city. The city itself seems to have a quite high impact in the model. What we are going to do with this variable depends highly what we want to predict. We should leave the variable in the model, if we are trying to predict prices in boston or seattle, but we would have to drop it to predict other cities.
